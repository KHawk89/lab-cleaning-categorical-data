```python
continous = pd.DataFrame()
discrete = pd.DataFrame()
categorical = pd.DataFrame()
time = pd.DataFrame()
def checkcatnum(df):
    for column in df:
        if df[column].dtype == 'int64':
            discrete[column] = df[column]
        elif df[column].dtype == 'float64':
            continous[column] = df[column]
        elif df[column].dtype == 'object':
            categorical[column] = df[column]
        else:
            time[column] = df[column]
    for col in discrete:
        sns.displot(data=discrete[col]) 
    for cols in continous:
        sns.displot(data=continous[cols],kind='ecdf')
    return plt.show()
checkcatnum(customer_df)
```

```python
def remove_outliers(df, threshold=1.5, in_columns=df.select_dtypes(np.number).columns, skip_columns=[]):
    for column in in_columns:
        if column not in skip_columns:
            upper = np.percentile(df[column],75)
            lower = np.percentile(df[column],25)
            iqr = upper - lower
            upper_limit = upper + threshold * iqr
            lower_limit = lower - threshold * iqr
            df = df[(df[column]>lower_limit) & (df[column]<upper_limit)]
    return df
```

```python
df2 = remove_outliers(df, threshold=4, in_columns=['customer_lifetime_value','income','total_claim_amount'])
df2
```
