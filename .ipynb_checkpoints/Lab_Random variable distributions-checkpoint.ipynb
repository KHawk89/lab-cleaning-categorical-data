```python
customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'])
customer_df.info()
```

```python
X_num = customer_df.select_dtypes(include = np.number)
X_num1 = X_num.copy()
```

```python
for col in X_num:
    sns.distplot(X_num[col])
    plt.show()
```

```python
def log_transfom_clean(x):
    x = np.log(x)
    if np.isfinite(x):
        return x
    else:
        return np.NAN
```

```python
for col in X_num:
    X_num[col] = list(map(log_transfom_clean, X_num[col]))
```

```python
for col in X_num:
    sns.distplot(X_num[col])
    plt.show()
```

```python
for col in X_num1:
    X_num1[col] = np.where(X_num1[col]<=0, X_num1[col].mean(), X_num1[col]) # X_num1[col].mean()  #np.NAN
    X_num1[col] = X_num1[col].fillna(X_num1[col].mean())
    transformed_data, _ci = stats.boxcox(X_num1[col])
    sns.distplot(transformed_data)
    plt.show()
```

### It seems that especially a boxcox transformation gives a slightly better result concerning the distribution.
